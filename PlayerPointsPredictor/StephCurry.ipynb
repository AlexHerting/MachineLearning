```python
import pandas as pd

# Load the CSV file
games = pd.read_csv('StephStats.csv', index_col="date")
```

```python
games.apply(pd.isnull).sum()/games.shape[0]
```

```python
#convert MP to float64
games['MP'] = games['MP'].apply(lambda x: int(x.split(':')[0]) + int(x.split(':')[1])/60 if isinstance(x, str) else x)
games['MP'] = games['MP'].astype('float64')
```

```python
games.dtypes
```

```python
#convert win/loss to float64
games['win_or_lose'] = games['win_or_lose'].apply(lambda x: 0 if x == 'L' else 1)

# Now 'win_or_lose' is in binary format (0 or 1), you can convert it to float64
games['win_or_lose'] = games['win_or_lose'].astype('float64')
```

```python
games.dtypes
```

```python
#age is converted to float 64
games['age'] = games['age'].apply(lambda x: float(x.split('-')[0]) if isinstance(x, str) else x)

# Now 'age' is in decimal format (age in years), you can convert it to float64
games['age'] = games['age'].astype('float64')
```

```python
games.dtypes
```

```python
games
```

```python
#home or away converted to float64
games['home_or_away'] = games['home_or_away'].apply(lambda x: 0 if x == '@' else 1)

# Now 'home_or_away' is in binary format (0 or 1), you can convert it to float64
games['home_or_away'] = games['home_or_away'].astype('float64')
```

```python
games.dtypes
```

```python
#convert our date to date-time for the ML
games.index = pd.to_datetime(games.index)
```

```python
games.index
```

```python
games = games.drop(['teams_rank', 'opp_rank'], axis=1)
```

```python
games
```

```python
games = games.dropna(subset=['game_number'])
```

```python
#no null data, we are all good to go
games.apply(pd.isnull).sum()/games.shape[0]
```

```python
#figure out what we want to predict
#create a target (what we want to predict)
#creates a colum with next games PTS
games.loc[:, "target"] = games["PTS"].shift(-1)
```

```python
games
```

```python
#removes the last row since we don't have the next games actual points
games = games.iloc[:-1,:].copy()
```

```python
games
```

```python
from sklearn.linear_model import Ridge
#the greater the penalty, the more overfitting is prevented
#We initialized the model for ML
reg = Ridge(alpha=.1)

#these values are used to predict next games points
predictors = ["game_number","age","home_or_away","win_or_lose","GS","MP","FG","FGA","3P","3PA","FT","FTA","ORB","DRB","TRB","AST","TOV","PF","PTS","GmSc","+/-"]
```

```python
#split the data into a training set and a testing set
#always want the past data to be the train data and the future data to be the test data
#everything up to 2020-12-25 is in the training set
train = games.loc[:"2020-12-25"]
```

```python
#everything after 2020-12-27 is in the testing set
test = games.loc["2020-12-27":]
```

```python
#fit our model to our predictors and try and predict our target
reg.fit(train[predictors], train["target"])
```

```python
#generate predictions on our test dataset
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
#need this for calculations
```

```python
#returns on average how many we were off by
mean_absolute_error(test["target"], predictions)
```

```python
#combine the actual values with the combined values to see where we have big differences
combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
combined.columns = ["actual", "predictions"]
```

```python
#functions to try more models easily
def create_predictions(predictors, core_weather, reg):
    #split the data into a training set and a testing set
    #always want the past data to be the train data and the future data to be the test data
    #everything up to 2020-12-25 is in the training set
    train = games.loc[:"2020-12-25"]
        
    #everything after 2020-12-27 is in the testing set
    test = games.loc["2020-12-27":]
    
    #fit our model to our predictors and try and predict our target
    reg.fit(train[predictors], train["target"])
    
    #generate predictions on our test dataset
    predictions = reg.predict(test[predictors])
    
    #returns on average how many we were off by
    error = mean_absolute_error(test["target"], predictions)
    
    #combine the actual values with the combined values to see where we have big differences
    combined = pd.concat([test["target"], pd.Series(predictions, index=test.index)], axis=1)
    combined.columns = ["actual", "predictions"]
    
    return error, combined
```
